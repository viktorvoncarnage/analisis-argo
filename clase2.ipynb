# PARTE 2: NumPy — El Array Numérico

````python
import numpy as np
print('NumPy', np.__version__)
````

### Arrays, propiedades y operaciones vectorizadas

````python
# Temperaturas superficiales del Mediterráneo (°C)
temps = np.array([22.1, 18.5, 25.3, 19.8, 23.7, 17.2, 24.5, 20.1])

print(f'Array:  {temps}')
print(f'Shape:  {temps.shape}  |  Dtype: {temps.dtype}  |  Ndim: {temps.ndim}')

# Operación vectorizada: sin bucle for
temps_f = (temps * 9/5) + 32
print(f'\nFahrenheit: {temps_f.round(1)}')

# Profundidades estándar de perfiles Argo
profundidades = np.linspace(0, 2000, 15)
print(f'\nProfundidades Argo (m): {profundidades}')
````

````python
temps = np.array([22.1, 18.5, 25.3, 19.8, 23.7])   # desde lista 
 
np.zeros(5)                  # [0. 0. 0. 0. 0.] 
np.ones(3)                   # [1. 1. 1.] 
np.linspace(0, 2000, 15)     # 15 profundidades entre 0 y 2000 m (Argo) 
np.arange(0, 100, 10)        # [0, 10, 20, ..., 90] 
 
# Datos de prueba 
rng = np.random.default_rng(42) 
temps_sim = rng.normal(20, 3, 1000)   # 1000 temperaturas simuladas
````

````python
temps_sim
````

### Máscaras booleanas — el mismo mecanismo que Pandas

````python
# La condición genera un array de True/False
mask = temps > 22
print(f'Máscara (>22°C):       {mask}')
print(f'Temperaturas cálidas:  {temps[mask]}')

# Forma abreviada
print(f'Forma abreviada:       {temps[temps > 22]}')

# Condición combinada: & para AND, | para OR (igual que en Pandas)
print(f'Entre 19 y 23°C:       {temps[(temps >= 19) & (temps <= 23)]}')
````

### Estadísticas y NaN

````python
print(f'Media:           {np.mean(temps):.2f} °C')
print(f'Std:             {np.std(temps):.2f} °C')
print(f'Rango:           {np.min(temps):.1f} – {np.max(temps):.1f} °C')
print(f'Percentil 25/75: {np.percentile(temps, 25):.1f} / {np.percentile(temps, 75):.1f} °C')
````

````python
# NaN: habitual en datos Argo (sensor caído, boya que no llegó a esa profundidad)
datos = np.array([22.1, np.nan, 25.3, np.nan, 23.7, 17.2])

print(f'Posiciones NaN:    {np.isnan(datos)}')
print(f'np.mean (con NaN): {np.mean(datos)}')           # → nan
print(f'np.nanmean:        {np.nanmean(datos):.2f}')    # correcto
print(f'Valores válidos:   {np.sum(~np.isnan(datos))} de {len(datos)}')
````

````python
~np.isnan(datos)
````

### ✏️ Ejercicio 2 — NumPy: estadísticas de salinidad

````python
sal = np.array([38.2, 37.8, np.nan, 38.9, 37.5, np.nan, 38.4, 39.1, np.nan, 37.9, 38.7])

# TU CÓDIGO:
# a) ¿Cuántos NaN hay?
# b) Media, máximo y percentil 90 ignorando NaN
# c) Valores con salinidad > 38.5 PSU
# d) ¿Cuántos valores son válidos?
nannum=np.sum(np.isnan(sal))
print('a)', nannum)
print('b) Media: ', np.nanmean(sal), ' PSU')
print('Máximo: ', np.nanmax(sal), ' PSU')
print('Percentil 90: ', np.nanpercentile(sal,90),' PSU')
print('c) Salididad alta: ', sal[sal>38.5],'PSU')
print('d) Valores válidos: ', len(sal)-nannum)
````

> 🤖 **Copilot tip:** *"Tengo un array NumPy con NaN. ¿Cómo cuento los NaN, calculo estadísticas ignorándolos y filtro por un umbral?"*

````python
# ✅ SOLUCIÓN
print(f'a) NaN: {np.sum(np.isnan(sal))}')
print(f'b) Media: {np.nanmean(sal):.3f} | Máx: {np.nanmax(sal):.3f} | P90: {np.nanpercentile(sal, 90):.3f} PSU')
print(f'c) > 38.5 PSU: {sal[sal > 38.5]}')
print(f'd) Válidos: {np.sum(~np.isnan(sal))} de {len(sal)}')
````

### 3.1 Cargar y explorar el dataset

### 3.2 Selección y filtrado

````python
# Filtrado simple
df_calido = df[df['temperature'] > 22]
print(f'Mediciones > 22°C: {len(df_calido):,} ({len(df_calido)/len(df)*100:.1f}%)')

# Condición múltiple — OJO: & con paréntesis, nunca and
df_sc = df[(df['temperature'] > 22) & (df['depth'] < 50)]
print(f'Superficial (< 50m) Y cálido (> 22°C): {len(df_sc):,}')
````

````python
# loc vs iloc
print('loc — por nombre:')
print(df.loc[0:3, 'temperature':'depth'])

print('\niloc — por posición:')
print(df.iloc[0:4, 4:8])
````

### 3.3 Valores nulos

````python
nulos = df.isnull().sum()
pct = (nulos / len(df) * 100).round(1)
print(pd.DataFrame({'Nulos': nulos, '%': pct})[nulos > 0])
````

````python
df_limpio = df.dropna(subset=['temperature', 'salinity']).copy()
print(f'Original: {len(df):,}  →  Limpio: {len(df_limpio):,}  (eliminadas: {len(df)-len(df_limpio):,})')
print('→ Los NaN suelen indicar que la boya no llegó a esa profundidad.')
````

# PARTE 4: Pandas II — Transformación

````python
# Fechas, nuevas columnas y apply()
df_limpio['date']  = pd.to_datetime(df_limpio['date'])
df_limpio['month'] = df_limpio['date'].dt.month
df_limpio['year']  = df_limpio['date'].dt.year
df_limpio['temp_f']     = df_limpio['temperature'] * 9/5 + 32
df_limpio['is_surface'] = df_limpio['depth'].apply(lambda d: d < 50)

def clasificar_profundidad(d):
    """Clasifica la profundidad según zonas oceánicas estándar."""
    if d < 10:     return 'superficie'
    elif d < 200:  return 'epipelágico'
    elif d < 1000: return 'mesopelágico'
    else:          return 'batipelágico'

df_limpio['zona'] = df_limpio['depth'].apply(clasificar_profundidad)

print('Distribución por zona:')
print(df_limpio['zona'].value_counts())
````

````python
# Agrupaciones: adelanto de lo que haremos en Clase 3
df_limpio.groupby('basin').agg(
    media_temp   = ('temperature', 'mean'),
    std_temp     = ('temperature', 'std'),
    media_sal    = ('salinity',    'mean'),
    n_mediciones = ('temperature', 'count')
).round(3).sort_values('media_temp', ascending=False)
````

````python
# Guardar el dataset limpio
out = os.path.join('data', 'argo_limpio.csv')
df_limpio.to_csv(out, index=False)
print(f'Guardado: {out}  ({len(df_limpio):,} filas)')
````
